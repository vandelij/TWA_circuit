```python
wall2 = np.loadtxt('DIIIDwall2.txt')
def plot_device(wall, idx):
    plt.figure(figsize=(12,12))
    plt.grid()
    R = wall[:,0]
    Z = wall[:, 1]
    print(f'X, Y:')
    print(R[idx], Z[idx])
    plt.plot(R,Z, marker='.', color='k')
    plt.scatter(R[idx], Z[idx], color='red', marker='o')
    plt.axis('equal')
    plt.ylim(0,1.5)
    plt.xlim(1,2.5)
```
