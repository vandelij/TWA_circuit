```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
npar_min = -20
npar_max = 20
num_npar=1000
npar_grid = np.linspace(npar_min, npar_max, num_npar)

def guassian(x, mean, sigma):
    return np.exp(-(x - mean)**2/sigma**2)

def normalized_npar_ideal(npar_array, ntarget, sigma):
    power1 = guassian(npar_array, mean=ntarget, sigma=sigma)
    power2 = guassian(npar_array, mean=-ntarget, sigma=sigma)
    total_power = (power1 + power2) / (np.max(power1 + power2))
    return total_power
    

```

```python
plt.figure(figsize=(10,4))
power = normalized_npar_ideal(npar_grid, ntarget=5, sigma=.8)
plt.plot(npar_grid, power, marker='.')
plt.grid()
plt.title(r'Ideal $n_{||}$ Spectrum')
plt.xlabel(r'$n_{||}$')
plt.ylabel('Power [a.u.]')
```
