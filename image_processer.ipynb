```python
import PIL.Image as Image
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
```

```python
image_path = '../van_zeland_2.png'

img = Image.open(image_path).convert("RGBA")

# new stuff

width, height = img.size

pixels = img.load()

# for x in range(13*width//30, 20*width//30):
#     for y in range(height//2, height):
#         i = width - x
#         j = height - y
#         pixels[i,j] = (0,0,0,255)





# for x in range(0, width):
#     for y in range(0, height):
#         if pixels[x,y] == (159, 0, 0, 255):
#             pixels[x,y] = (255, 255, 255, 255)

for x in range(13*width//30, 20*width//30):
    for y in range(height//2, height):
        i = width - x
        j = height - y
        if (pixels[i,j][0] > pixels[i,j][1]) and (pixels[i,j][0] > pixels[i,j][2]):
            pixels[i,j] = (255, 255, 255, 255) 
```

```python
# # Save the updated image
output_path = "output.png"  # Replace with your desired output path
img.save(output_path)

print(f"Color replaced and saved to {output_path}")
```
